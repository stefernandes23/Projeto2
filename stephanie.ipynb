```python
import pandas as pd
from datetime import datetime
import csv
dados = dict()
dadosQuest = dict()

class Pessoa:
    
    def __init__(self, idade, genero, estado, cidade, data):
        idade = ''
        genero = ''
        estado = ''
        cidade = ''
        data = ''
        dados = []
        
    def entrevistado(self, idade, genero, estado, cidade, data):
        self.idade = idade
        if (idade > 0) and (idade < 120): #delimitando idade
            self.genero = genero
            if (genero == 'f') or (genero == 'm'): #delimitando forma de escrever gênero
                if (genero == 'f'):
                    genero = ('feminino')
                elif (genero == 'm'):
                    genero = ('masculino')
                else:
                    print('Digite corretamente o seu gênero!')
                self.estado = estado
                self.cidade = cidade
                self.data = data
            self.dados = {'idade':idade, 'genero':genero, 'estado':estado, 'cidade':cidade, 'data': data}
            print(dados)


class Questionario:
    
    def __init__(self, pergunta1, pergunta2, pergunta3, pergunta4):
        pergunta1 = ''
        pergunta2 = ''
        pergunta3 = ''
        pergunta4 = ''
        dadosQuest = []
    
    def pergunta (self, pergunta1, pergunta2, pergunta3, pergunta4):
        self.pergunta1 = pergunta1
        if pergunta1 == 1:
            pergunta1 = 'Ensino Fundamental Incompleto'
        elif pergunta1 == 2:
            pergunta1 = 'Ensino Fundamental Completo'
        elif pergunta1 == 3:
            pergunta1 = 'Ensino Médio Incompleto'
        elif pergunta1 == 4:
            pergunta1 = 'Ensino Médio Completo'
        elif pergunta1 == 5:
            pergunta1 = 'Ensino Superior Incompleto'
        elif pergunta1 == 6:
            pergunta1 = 'Ensino Superior Completo'
        elif pergunta1 == 7:
            pergunta1 = 'Pós-Graduação Completa'
        self.pergunta2 = pergunta2
        if pergunta2 == 1:
            pergunta2 = 'até R$ 1.200,00'
        elif pergunta2 == 2:
            pergunta2 = 'R$ 1.201,00 - R$ 2.000,00'
        elif pergunta2 == 3:
            pergunta2 = 'R$ 2.001,00 - R$ 4.300,00'
        elif pergunta2 == 4:
            pergunta2 = 'R$ 4.301,00 - R$ 8.000,00'
        elif pergunta2 == 5:
            pergunta2 = '5- 8.001,00 - R$ 15.000,00'
        elif pergunta2 == 6:
            pergunta2 = 'maior que F$ 15.001,00'
        self.pergunta3 = pergunta3
        if pergunta3 == 1:
            pergunta3 = 'Piorou muito'
        elif pergunta3 == 2:
            pergunta3 = 'Piorou um pouco'
        elif pergunta3== 3:
            pergunta3 = 'Nem piorou, nem melhorou'
        elif pergunta3== 4:
            pergunta3 = 'Melhorou um pouco'
        elif pergunta3== 5:
            pergunta3 = 'Melhorou muito'
        self.pergunta4 = pergunta4
        if pergunta4 == 'sim' or 's':
            pergunta4 = input('Procurou alguma ajuda psicológica?\nDigite[S/N]:')
            if pergunta4 == 'sim' or 's':
                pergunta4 = 'procurou ajuda psicológica'
            else:
                pergunta4 = 'sim e não procurou ajuda psicológica'
        self.dadosQuest = {'Grau de escolaridade': pergunta1, 'Renda financeira': pergunta2, 'Percepção sobre saúde mental': pergunta3, 'Ajuda psicológica': pergunta4}
        print(dadosQuest)
        dados.update(dadosQuest)
        

lista_entrevistados = []
loop = True
from datetime import datetime
while loop: 
    lista = []
    cont = 0
    idade = int(input('Digite a sua idade: '))
    genero = input('Digite o seu gênero [F/M]: ').lower()
    estado = input('Digite o estado em que mora: ').lower()
    cidade = input('Digite a cidade em que mora: ').lower()
    data = datetime.now().strftime('%Y-%m-%d %H:%M')
    dadosPessoa = Pessoa (idade, genero, estado, cidade, data)
    dadosPessoa.entrevistado(idade, genero, estado, cidade, data)
    cont += 1
    
    print("Vamos ao questionário:")
    pergunta1 = int(input('Qual o seu grau de escolaridade?\n1- Ensino Fundamental Incompleto;\n2- Ensino Fundamental Completo;\n3- Ensino Médio Incompleto\n4- Ensino Médio Completo;\n5- Ensino Superior Incompleto;\n6- Ensino Superior Completo;\n7-Pós-Graduação completa;\n\nDigite:'))

    pergunta2 = int(input('Renda financeira:\n1- até R$ 1.200,00\n2- R$ 1.201,00 - R$ 2.000,00;\n3- R$ 2.001,00 - R$ 4.300,00;\n4- R$ 4.301,00 - R$ 8.000,00;\n5- 8.001,00 - R$ 15.000,00;\n6- maior que 15.001,00\nDigite sua resposta: '))

    pergunta3 = int(input('Qual a sua percepção sobre a sua saúde mental durante a pandemia nos quesitos: sintomas de depressão, ansiedade, stress, síndrome do pânico e/ ou compulsões,\n1- Piorou muito\n2- Piorou um pouco\n3- Nem piorou, nem melhorou\n4- Melhorou um pouco\n5-Melhorou muito\nDigite sua resposta: '))

    pergunta4 = input('Você tem sentido tristeza, melancolia, desespero, desesperança? Se sim, procurou alguma ajuda psicológica? \nDigite sua resposta[S/N]: ').lower
    pergunta = Questionario(pergunta1, pergunta2, pergunta3, pergunta4)

    pergunta.pergunta(pergunta1, pergunta2, pergunta3, pergunta4)
    cont += 1

    dadosPessoa.dados.update(pergunta.dadosQuest)
    
    field_names = ['idade','genero','estado', 'cidade','Grau de escolaridade', 'Renda financeira', 'Percepção sobre saúde mental', 'Ajuda psicológica','data']

    lista_entrevistados.append(dadosPessoa.dados)

    with open ('Test.csv', 'w', newline = '' ) as csvfile:
        escritor = csv.DictWriter(csvfile, fieldnames = field_names)
        escritor.writeheader()
        escritor.writerows(lista_entrevistados)
    
    continuar = input('Deseja continuar? ')
    if continuar == 'n':
        loop = False

print(dadosPessoa.dados)
print (lista_entrevistados)

    

```
